```python
#Data Aggregation
import pandas as pd
from numpy.random import seed
from numpy.random import rand
from numpy.random import randint
import numpy as np
```

```python
seed(42) #ê³ ì • ëžœë¤ ë‚œìˆ˜
```

```python
df = pd.DataFrame(
    {'Weather' : ['cold', 'hot', 'cold', 'hot','cold', 'hot', 'cold'],
   'Food' : ['soup', 'soup', 'icecream', 'chocolate','icecream', 'icecream', 'soup'],
   'Price' : 10 * rand(7), 
    'Number' : randint(1, 9)}
)
print(df)
```

```python
weather_group=df.groupby('Weather') #ê·¸ë£¹í™”
#weather_group ->DATAFRAMEGROUPBY íƒ€ìž…ìž„ 
i=0
for name, group in weather_group: 
    i=i+1
    print("Group",i,name) 
    print(group)
    
#ê° ë°ì´í„°ëŠ” dataframeìœ¼ë¡œ ë¨
```

```python
#ê° ê·¸ë£¹ ì²«ë²ˆì§¸í–‰, ë§ˆì§€ë§‰ í–‰, í‰ê·  ì¶œë ¥
print("Weather group first\n", weather_group.first())
print("Weather group last\n", weather_group.last())
print("Weather group mean\n", weather_group.mean())
```

```python
wf_group=df.groupby(['Weather','Food'])
#print("WF Groups",wf_group.groups)
#í…ŒìŠ¤íŠ¸ë¡œ ië¥¼ ë‹¤ì‹œ 0ë¶€í„°í•´ì„œ ë³¼ë•Œ , enumerateë¡œ ê³„ì‚°
for i,(name, group) in enumerate(weather_group):
    i=i+1
    print("Group",i,name) 
    print(group)
```

```python
print("WF Aggregated\n",wf_group.agg([np.mean,np.median]))
#agg() = apply() ì ìš©
#np.mean í•¨ìˆ˜ / np.mean() í•¨ìˆ˜ê°’
#agg(í•¨ìˆ˜) : í•¨ìˆ˜ë¥¼ ë°›ëŠ” í•¨ìˆ˜ / í•œë²ˆì— ì—¬ëŸ¬ê°œ ë°›ìŒ
```
